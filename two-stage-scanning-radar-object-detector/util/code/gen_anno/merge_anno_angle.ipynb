```python
import os
import json


bag1_anno = '/data2/itri/DCV/single_sensor_data/bag1/_front_angle/annotations'
bag2_anno = '/data2/itri/DCV/single_sensor_data/bag2/_front_angle/annotations'


bag1_all = {}
bag2_all = {}

split = ['train', 'val', 'test']
```

```python
for s in split:
	with open(os.path.join(bag2_anno, s+'.json'), 'r') as f:
		data = json.load(f)
		bag2_all[s] = data
		# print(data)

for s in split:
	with open(os.path.join(bag1_anno, s+'.json'), 'r') as f:
		data = json.load(f)
		bag1_all[s] = data
		# print(data)
```

```python
for k, v in bag1_all.items():
	print(v['images'][:4])
	print()
```

```python
for k, v in bag1_all.items():
	img_shift = v['images'][-1]['id']
	anno_shift = v['annotations'][-1]['id']

	for d in bag1_all[k]['images']:
		d.update({'file_name': 'bag1/radar_cart_train_test_split/'+d['file_name']})

	for d in bag2_all[k]['images']:
		d.update({'file_name': 'bag2/radar_cart_train_test_split/'+d['file_name'],
							'id': d['id']+img_shift})

	for d in bag2_all[k]['annotations']:
		d.update({'id': d['id']+anno_shift,
							'image_id': d['image_id']+img_shift})
```

```python
bag2_all['test']['images']
```

```python
from collections import defaultdict
out = defaultdict(dict)

for k in bag1_all:
	for kk in bag1_all[k]:
		out[k][kk] = bag1_all[k][kk] + bag2_all[k][kk]
```

```python
out['train']['annotations'][-1]
```

```python
for k, v in out.items():
	with open('{}.json'.format(k), "w") as f:
		json.dump(v, f, indent = 2)
```
