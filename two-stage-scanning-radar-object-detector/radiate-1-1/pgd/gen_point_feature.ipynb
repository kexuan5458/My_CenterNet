```python
import numpy as np
import matplotlib.pyplot as plt

def ret_normalize(x):
		return (x - np.mean(x)) / (np.std(x))

def ret_x_y_range():
		shape = (1152, 1152)

		x = np.linspace(-100, 100, shape[1])
		y = np.linspace(100, -100, shape[1])

		X, Y = np.meshgrid(x, y)

		ran = np.sqrt(X**2 + Y**2)
		ang = np.arctan2(X, Y) % (2*np.pi)

		return list(map(ret_normalize, (X, Y, ran, ang)))

ret_x_y_range()
```
